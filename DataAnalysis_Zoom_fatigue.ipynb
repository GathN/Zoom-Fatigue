Dapat terlihat, data yang seharusnya numerik ternyata masih berbentuk objek, karena data yang berisi desimal, diisi dengan (,) dan bukan (.)

```python
df["avg_usage"] = pd.to_numeric(df["avg_usage"].str.replace(',', '.').astype(float))
df["screen"] = pd.to_numeric(df["screen"].str.replace(',', '.').astype(float))
df["limit"] = pd.to_numeric(df["limit"].str.replace(',', '.').astype(float))
```

Mengecek apakah terdapat nilai null dalam dataset

```python
df.isna().sum()
```

##Melakukan Normalisasi Data

```python
normalized_df = df
```

```python
normalized_df.head()
```

Dilakukan normalisasi min-max untuk semua data numerik, sementara data kategorikal tidak dilakukan normalisasi

```python
normalized_df.iloc[:,10:]=(normalized_df.iloc[:,10:]-normalized_df.iloc[:,10:].min())/(normalized_df.iloc[:,10:].max() - normalized_df.iloc[:,10:].min())
normalized_df.iloc[:,3:9]=(normalized_df.iloc[:,3:9]-normalized_df.iloc[:,3:9].min())/(normalized_df.iloc[:,3:9].max() - normalized_df.iloc[:,3:9].min())
normalized_df.iloc[:,0:2]=(normalized_df.iloc[:,0:2]-normalized_df.iloc[:,0:2].min())/(normalized_df.iloc[:,0:2].max() - normalized_df.iloc[:,0:2].min())
```

```python
normalized_df.head()
```

##Persebaran Data (Histogram)


Melakukan analisis dan visualisasi untuk persebaran data setiap kolom.

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_melted = normalized_df.melt(var_name='column')
plot = sns.FacetGrid(df_melted, col='column', col_wrap=5, sharex=False)


plot.map(plt.hist, 'value')
for ax in plot.axes.flat:
  if(ax == 3):
    plt.sca(3)
    plot.set(xticks=normalized_df.iloc[:, 2])
  if(ax == 10):
    plt.sca(10)
    plot.set(xticks=normalized_df.iloc[:, 9])
  ax.set_xticklabels(ax.get_xticklabels(), rotation = 45)
plt.tight_layout()
```

##Correlation Map

###Mencari nilai korelasi antara dua kolom/fitur data. <br>
* Bisa terlihat nilai korelasi tinggi merupakan keterkaitan apabila salah satu data naik saat data lain naik. <br>
* Sedangkan ketika nilai korelasi rendah(negatif), terjadi ketika salah satu data bernilai turun saat data yang lain naik, dan sebaliknya.<br>
* Sementara untuk nilai korelasi 0 atau mendekati 0, tidak ditemukan/ditemukan sedikit korelasi.

```python
plt.figure(figsize=(20,15))
sns.heatmap(df.corr())
```
